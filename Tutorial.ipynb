## Draw Proteomics fractions vs. growth rate

```python
file_path = './data/Balakrishnan2022_proteomics_colMetadata.csv'
metadata = pd.read_csv(file_path, encoding='mac_roman')
```

```python
metadata.columns=['number','sample', 'perturbation_condition', 'g_rates', 'description']
```

```python
metadata
```

```python
file_path = './data/Balakrishnan2022_proteomics_fractions.csv'
proteomics = pd.read_csv(file_path, encoding='mac_roman')
```

```python
proteomics.rename(columns={proteomics.columns[0]: "gene" }, inplace = True)
```

```python
proteomics
```

```python
#extract proteome fractions for proteins belonging to a target class
target = 'H'
res_C = []
res_R = []
for i in range(3738):
    gene_str = proteomics['gene'][i]
    current = Ecoli_genes['cog_functional_cat'][i]
    if target == current:
        gene_index = i
        Clim_concentration=[]
        Rlim_concentration=[]
        for j in range(len(metadata['sample'])):
            if metadata['perturbation_condition'][j]=='Clim':
                Clim_concentration.append(proteomics[metadata['sample'][j]][gene_index])
            if metadata['perturbation_condition'][j]=='Rlim':
                Rlim_concentration.append(proteomics[metadata['sample'][j]][gene_index])
        res_C.append(Clim_concentration)
        res_R.append(Rlim_concentration)
```

```python
#saving growth rates for plotting
C_growth_rates = []
R_growth_rates = []
for i in range(len(metadata)):
    if metadata['perturbation_condition'][i]=='Clim':
        C_growth_rates.append(float(metadata['g_rates'][i].replace(',','.')))
    if metadata['perturbation_condition'][i]=='Rlim':
        R_growth_rates.append(float(metadata['g_rates'][i].replace(',','.')))

C_growth_rates = np.array(C_growth_rates)
R_growth_rates = np.array(R_growth_rates)
```

```python
# avg. over the protein class
res_C = np.array(res_C)
#remove the NaN values from the array
res_C = res_C[~np.isnan(res_C).any(axis=1)]
avg_C = np.mean(res_C, axis=0)
err_C = np.std(res_C, axis=0)/len(res_C)
```

```python
# avg. over the protein class
res_R = np.array(res_R)
res_R = res_R[~np.isnan(res_R).any(axis=1)]
avg_R = np.mean(res_R, axis=0)
err_R = np.std(res_R, axis=0)/len(res_C)
```

```python
#linear fit
mC, bC = np.polyfit(C_growth_rates, avg_C, 1)
mR, bR = np.polyfit(R_growth_rates, avg_R, 1)
```

```python
#plotting
plt.errorbar(x=C_growth_rates, y=avg_C, yerr=err_C, fmt='o', color='tab:red',capsize=2)
plt.plot(np.linspace(0,1,20), mC*np.linspace(0,1,20)+bC, color='tab:red', linestyle='dashed',label='C limitation')
plt.errorbar(x=R_growth_rates, y=avg_R, yerr=err_R, fmt='o', color='tab:green',capsize=2)
plt.plot(np.linspace(0,1,20), mR*np.linspace(0,1,20)+bR, color='tab:green',linestyle='dashed', label='R limitation')
plt.xlabel(r'Growth rate ($\lambda$)', size=15)
plt.ylabel(r'Fraction of proteome ($\phi$)', size=15)
plt.title('Functional category: '+target, size=20)
plt.legend()
plt.show()
```
